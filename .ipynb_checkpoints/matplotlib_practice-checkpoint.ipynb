```python
%matplotlib inline
```

```python
# Import Dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create plot and add a label for the legend.
plt.plot(x_axis, y_axis, label = 'Boston')
# Create labels for x and y axis
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set y limit between 0 and 45
plt.ylim(0,45)
# Create title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Changing markers, line colour and adding grid
plt.plot(x_axis, y_axis, marker = "*", color="blue", linewidth = 2, label="Boston")
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.ylim(0,45)
plt.title("PyBer Fare by Month")
plt.grid()
plt.legend()
```

```python
# Creating Chart using OO Approach
fig, ax = plt.subplots(figsize=(13,4))
ax.set_ylim(5,45)
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
ax.set_title("PyBer Fare by Month")
ax.grid()
ax.plot(x_axis, y_axis, label="Boston", marker="D", color="Green",linewidth=2)
ax.legend()
```

```python
# Create Bar Chart
plt.bar(x_axis, y_axis, color="green", label="Boston")
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
plt.barh(x_axis, y_axis)
```

```python
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
# Skill Drill
plt.barh(x_axis, y_axis, color = "magenta", label = "Boston")
plt.title("PyBer Fare by Month")
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.gca().invert_yaxis()
plt.legend()
```

# Object Oriented Charts

```python
# Skill Drill
fig, ax = plt.subplots(figsize=(16,3))
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.set_title("PyBer Fare by Month")
ax.barh(x_axis, y_axis, label="Chicago", color="cyan")
plt.gca().invert_yaxis()
ax.legend()
```

# Skill Drill

```python
fig, ax = plt.subplots()
ax.set_title("PyBer Fare by Month")
ax.set_xlim(0,50)
ax.scatter(y_axis, x_axis, s = [i *5 for i in y_axis], color = "xkcd:sky blue", alpha = 0.2, 
           edgecolors="black", linewidth = 2, label="Boston")
plt.gca().invert_yaxis()
ax.legend()
```

# Pie Chart

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
plt.subplots(figsize =(8,8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

# Skill Drill
# Creating pie using Object Oriented Approach

```python
fig, ax= plt.subplots(figsize=(8,8))
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
ax.pie(y_axis, labels=x_axis, 
       autopct='%.1f%%', explode = explode_values, 
       shadow=True, startangle=-95,
      colors = colors)
plt.gca().invert_yaxis()
plt.show
```

# Add Error lines to Line Chart

```python
import statistics
```

```python
# Get standard deviation of the values in the y-axis.
stdev = statistics.stdev(y_axis)
stdev
```

```python
plt.errorbar(x_axis, y_axis, yerr=stdev)
```

```python
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

#Errorbar using Object Oriented Method

```python
fig, ax = plt.subplots(figsize=(13,1))
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize = 3)
plt.show()
```

```python
plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
# Changing major Ticks
```

```python
import numpy as np
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0,51, step=5.0))
plt.gca().invert_yaxis()
```

```python
from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
